```python
import pandas as pd

df = pd.read_csv("findex_clean.csv").dropna()
cols = ['Account','Savings','Debit','Credit','DigitalPay']
# remove all the '%' and treat it numeric
for col in cols:
    df[col] = df[col].str.replace('%','').astype(float)
df.head()
```

```python
import plotly.express as px

# digital payment adoption by country and year
fig = px.line(
    df,
    x="year",
    y="DigitalPay",
    color="country"
)

fig.show()

# account ownership
fig2 = px.scatter(
    df,
    x="Account",
    y="DigitalPay",
    color="country",
    animation_frame='year'
)
fig2.show()

# credit + debit
plot_df = df.melt(id_vars=['country','year'],
                  value_vars=['Debit','Credit'],
                  var_name='card_type',
                  value_name='ownership')

# remove 2024 cuz data look sus
plot_df = plot_df[plot_df['year']!=2024]

fig3 = px.bar(
    plot_df,
    x='country',
    y='ownership',
    color='card_type',
    barmode='group',
    animation_frame='year'
)

fig3.show()
```

```python
from sklearn.preprocessing import StandardScaler

metrics = ['Account','Debit','Credit','DigitalPay']

latest = df[df['year']==2021].copy()

scaler = StandardScaler()

latest[metrics] = scaler.fit_transform(latest[metrics])

latest['score'] = latest[metrics].mean(axis=1)

px.bar(
    latest.sort_values('score'),
    x='score',
    y='country',
    orientation='h'
)
```

```python
# correlation matrix
core_df = df[metrics]
corr = core_df.corr()

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

sns.heatmap(
    corr,
    annot=True,
    cmap='RdBu_r',
    center=0,
    fmt='.2f'
)

plt.title("Correlation Matrix")
plt.show()
```

Debit card ownership and account ownership appear more strongly associated with digital payment adoption than credit card ownership.

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# make new df on the growth
c = list(set(df['country'].values))
print(c)

df_growth = pd.DataFrame(columns=['country','growth'])
ct = []
diff = []
for i in c:
    tem1 = int(df[(df['country']==i) & (df['year']==2014)]['DigitalPay'].iloc[0])
    tem2 = int(df[(df['country']==i) & (df['year']==2024)]['DigitalPay'].iloc[0])
    ct.append(i)
    diff.append(tem2-tem1)
df_growth['country'] = ct
df_growth['growth'] = diff
```
